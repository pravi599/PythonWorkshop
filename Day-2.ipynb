```python
# Sum of factors

n = int(input('Enter number:'))
f_s = 0
for i in range(1,n):
    if n % i == 0:
        f_s+=i #f_s = f_s+i
        print(i,end=' \n')
print('Sum of factors is:',f_s)
```

```python
# To check perfect number

n = int(input('Enter number:'))
f_s = 0
for i in range(1,n):
    if n % i == 0:
        f_s+=i #f_s = f_s+i
        print(i,end=' \n')
print('Sum of factors is:',f_s)
if f_s == n:
    print('Perfect number')
```

```python
# To check perfect number or not

n = int(input('Enter number:'))
f_s = 0
for i in range(1,n):
    if n % i == 0:
        f_s+=i #f_s = f_s+i
        print(i,end=' \n')
print('Sum of factors is:',f_s)
if f_s == n:
    print('Perfect number')
else:
    print("Not a perfect number")

```

```python
#2.required or positional arg 

def required(a,b,c=30):
    return a*b*c
required(2,3)
```

```python
#3.Variable length arg

def Var_len(a,*b):
    return a+sum(b)
Var_len(4,5,6,7,7,7)
```

```python
def Is_prime(n):
    if n==1:
        return False
    cnt = 0
    for i in range(2,n):
        if n%i==0:
            cnt+=1
            break
    if cnt == 0:
        return True
    else:
        return False
Is_prime(5)    
```

```python
def prime_gen(s,e):
    for i in range(s,e+1):
        if Is_prime(i):
            print(i)
s=int(input('Enter start value:'))
e=int(input('Enter end value:'))
prime_gen(s,e)
            
        
            
            
```
